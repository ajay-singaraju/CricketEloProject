```python
import pandas as pd
from datetime import timedelta

for index, row in all_matches.iloc[::2].iterrows():  # Iterate through every other row
    # Accessing row data
    row_team = row['Team']
    row_opponent = row['Opposition']
    row_result = row['Result']
    row_date = row['Start Date']
    row_ground = row['Ground']  # Assign the ground for the match

    if row_result == 'won':
      result = 1.0
      opp_result = 0.0
    elif row_result == 'lost':
      result = 0.0
      opp_result = 1.0
    elif row_result == 'draw':
      result = 0.5
      opp_result = 0.5

    team_rows = elo_df[elo_df['Team'] == row_team]
    most_recent_team_row = team_rows.sort_values(by='Date', ascending=False).iloc[0]
    team_recent_elo = most_recent_team_row['ELO']

    opponent_rows = elo_df[elo_df['Team'] == row_opponent]
    most_recent_opponent_row = opponent_rows.sort_values(by='Date', ascending=False).iloc[0]
    opponent_recent_elo = most_recent_opponent_row['ELO']

    # Check if the home team is playing at home
    home_team_row = home_teams[home_teams['Ground'] == row_ground]

    if not home_team_row.empty:
        home_team = home_team_row.iloc[0]['Team']  # Extract the home team name

        if home_team == row_team:
            EO_team = 1 / (1 + 10**((team_recent_elo - opponent_recent_elo + 64) / 400))
            EO_opposition = 1 / (1 + 10**((team_recent_elo - opponent_recent_elo) / 400))
        elif home_team == row_opponent:
            EO_team = 1 / (1 + 10**((opponent_recent_elo - team_recent_elo + 64) / 400))
            EO_opposition = 1 / (1 + 10**((opponent_recent_elo - team_recent_elo) / 400))
    else:
        # Handle the case where the home team is not found (e.g., neutral venue)
        EO_team = 1 / (1 + 10**((team_recent_elo - opponent_recent_elo) / 400))
        EO_opposition = 1 / (1 + 10**((opponent_recent_elo - team_recent_elo) / 400))

    team_post_elo = team_recent_elo + 22 * (result - EO_team)
    opponent_post_elo = opponent_recent_elo + 22 * (opp_result - EO_opposition)

    new_row_team = pd.DataFrame({'Team': [row_team], 'ELO': [team_post_elo], 'Date': [row_date + timedelta(days=5)]})
    elo_df = pd.concat([elo_df, new_row_team], ignore_index=True)

    new_row_opponent = pd.DataFrame({'Team': [row_opponent], 'ELO': [opponent_post_elo], 'Date': [row_date + timedelta(days=5)]})
    elo_df = pd.concat([elo_df, new_row_opponent], ignore_index=True)

    print(elo_df)
```
